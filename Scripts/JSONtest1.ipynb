```python
# Creating our own JSON traffic scenario from scratch
# Attempted straightaway
import json
```

```python
# Data to be written
name = "tfrecord_00001-of-10000-429.json"

# Initialize lane_dict with proper structure
lane_dict = {}
x = 350.00000
start_y = 10000.0000
for i in range(200):
    lane_dict[str(i)] = {"x": x + i * 0.005, "y": start_y + i * 0.05}

lane_dict_2 = {}
x_2 = 400.00000
start_y = 10000.0000
for i in range(1000):
    lane_dict_2[str(i)] = {"x": x_2 + i * 0.005, "y": start_y + i * 0.05}

lane_dict_3 = {}
x_3 = 1000.00000
start_y = -10000.0000
for i in range(200):
    lane_dict_3[str(i)] = {"x": x_3 + i * 0.005, "y": start_y + i * 0.05}

lane_dict_4 = {}
x_4 = 1100.00000
start_y = -10000.0000
for i in range(1000):
    lane_dict_4[str(i)] = {"x": x_4 + i * 0.005, "y": start_y + i * 0.05}

# Define roads_dictionary with correct structure
roads_dictionary = {
    "0": {
        "geometry": lane_dict,
        "type": "road_edge"
    },
    "1": {
        "geometry": lane_dict_2,
        "type": "road_edge"
    },
    "2": {
        "geometry": lane_dict_3,
        "type": "road_edge"
    },
    "3": {
        "geometry": lane_dict_4,
        "type": "road_edge"
    }
}

# Define the main dictionary
dictionary = {
    "name": name,
    "objects": {},
    "roads": roads_dictionary,
    "tl_states": {}
}

# Print the dictionary
print(dictionary)
```

```python

# Serializing json
json_object = json.dumps(dictionary, indent=4)
 
# Writing to sample.json
with open(name, "w") as outfile:
    outfile.write(json_object)
```
